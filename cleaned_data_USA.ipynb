```python
import pandas as pd

# Wczytanie danych
df = pd.read_csv("/Users/mariolaczajkowska/anaconda_projects/python/project_mcz2/downloads/AB_US_2023.csv", low_memory=False)

# upper columns headers
df.columns = df.columns.str.upper()

# cleaning data (null and exluded columns)

df_cleaned = df.dropna(subset=['PRICE', 'NAME', 'CITY', 'NEIGHBOURHOOD_GROUP', 'LATITUDE', 'LONGITUDE']).copy() \
               .drop(columns=['NEIGHBOURHOOD', 'AVAILABILITY_365', 'ID', 'HOST_ID','LAST_REVIEW', 'REVIEWS_PER_MONTH', 'CALCULATED_HOST_LISTINGS_COUNT' ])

df_cleaned.insert(1, 'HOST_IDENTITY_VERIFIED', 'No information')
df_cleaned.insert(8, 'CANCELLATION_POLICY', 'No information')
df_cleaned.insert(7, 'INSTANT_BOOKABLE', 'No information')
df_cleaned.insert(10, 'REVIEW_RATE_NUMBER', 'No information')
df_cleaned.insert(11, 'CURRENCY', 'USD')
df_cleaned.insert(14, 'HOUSE_RULES', 'No information')
df_cleaned.insert(3, 'SERVICE_FEE', 'No information')

negative_values = df_cleaned[df_cleaned['MINIMUM_NIGHTS'] < 0]
df_cleaned = df_cleaned.drop(df_cleaned[df_cleaned['MINIMUM_NIGHTS'] < 0].index)

df_cleaned.rename(columns={
    'LATITUDE': 'LAT',
    'LONGITUDE' : 'LONG'
}, inplace=True)

df_cleaned['LAT'] = pd.to_numeric(df_cleaned['LAT'], errors='coerce')
df_cleaned['LONG'] = pd.to_numeric(df_cleaned['LONG'], errors='coerce')
df_cleaned = df_cleaned.dropna(subset=['LAT', 'LONG'])
# fill missing values
df_cleaned['CITY'] = df_cleaned['CITY'].replace('New York City', 'New York')

df_cleaned['HOST_NAME'] = df_cleaned['HOST_NAME'].fillna('Prefer not to say')
negative_values = df_cleaned[df_cleaned['MINIMUM_NIGHTS'] < 0]
df_cleaned = df_cleaned.drop(df_cleaned[df_cleaned['MINIMUM_NIGHTS'] < 0].index)
df_cleaned['MINIMUM_NIGHTS'] = df_cleaned['MINIMUM_NIGHTS'].replace('No minimum stay required', 0)
df_cleaned['MINIMUM_NIGHTS'] = pd.to_numeric(df_cleaned['MINIMUM_NIGHTS'], errors='coerce')  # Ensure numeric type
df_cleaned['NUMBER_OF_REVIEWS'] = df_cleaned['NUMBER_OF_REVIEWS'].fillna(0).astype(int)
df_cleaned['CITY'] = df_cleaned['CITY'].astype(str)
df_cleaned['PRICE'] = df_cleaned['PRICE'].astype(int)
df_cleaned = df_cleaned[df_cleaned['PRICE'] <= 2000]

new_order = ["NAME", "HOST_IDENTITY_VERIFIED", "HOST_NAME", "CITY", "NEIGHBOURHOOD_GROUP", "LAT",	"LONG", "INSTANT_BOOKABLE", "CANCELLATION_POLICY",	"ROOM_TYPE", "PRICE", "SERVICE_FEE", "CURRENCY", "MINIMUM_NIGHTS", "NUMBER_OF_REVIEWS", "REVIEW_RATE_NUMBER", "HOUSE_RULES"]

df_order = df_cleaned[new_order]

df_order.to_csv("downloads/cleaned_data_USA.csv", index=False)

print(df_order)
```
